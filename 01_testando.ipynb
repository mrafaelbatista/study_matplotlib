# 01 - Testando o Matplotlib

## Criando o primeiro gr√°fico

```python
import matplotlib.pyplot as plt
```

```python
squares = [1, 4, 9, 16, 25]
```

```python
plt.plot(squares)
plt.show()
```

### Alterando o r√≥tulo e a espessura

```python
# Aumenta a espessura da linha
plt.plot(squares, linewidth=5)

# Define o t√≠tulo do gr√°fico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos r√≥rulos das marca√ß√µes
plt.tick_params(axis='both', labelsize=4)

plt.show()
```

## Melhorando o gr√°fico

```python
# Definindo novos valores

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25] # repeti os valores, n√£o precisa, mas fica leg√≠vel
```

```python
# Aumenta a espessura da linha
plt.plot(input_values, squares, linewidth=5)

# Define o t√≠tulo do gr√°fico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos r√≥rulos das marca√ß√µes
plt.tick_params(axis='both', labelsize=4)

plt.show()
```
