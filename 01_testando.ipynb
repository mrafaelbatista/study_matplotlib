# 01 - Testando o Matplotlib

## Criando o primeiro gráfico

```python
import matplotlib.pyplot as plt
```

```python
squares = [1, 4, 9, 16, 25]
```

```python
plt.plot(squares)
plt.show()
```

### Alterando o rótulo e a espessura

```python
# Aumenta a espessura da linha
plt.plot(squares, linewidth=5)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', labelsize=4)

plt.show()
```

## Melhorando o gráfico

```python
# Definindo novos valores

input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25] # repeti os valores, não precisa, mas fica legível
```

```python
# Aumenta a espessura da linha
plt.plot(input_values, squares, linewidth=5)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', labelsize=4)

plt.show()
```
