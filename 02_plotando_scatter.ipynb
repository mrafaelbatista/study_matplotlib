# Plotando e estilizando pontos individuais com scatter()

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(2, 4)
plt.show()
```

```python
# Deixando mais interessante
plt.scatter(2, 4, s=200)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
```

## Plotando vários pontos

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

# Deixando mais interessante
plt.scatter(x_values, y_values, s=100)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Deixando mais interessante
plt.scatter(x_values, y_values, edgecolor='none', s=100)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Deixando mais interessante
plt.scatter(x_values, y_values, c='red', edgecolor='none', s=100)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Deixando mais interessante
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=100)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
```

```python
# Deixando mais interessante
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=100)

# Define o título do gráfico e nomeia os eixos
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Values", fontsize=14)

#Define o tamanho dos rórulos das marcações
plt.tick_params(axis='both', which='major', labelsize=14)
plt.savefig('squares_plot.png', bbox_inches='tight')
```
