# RandomWalk()

Random walk é utilizado para analisar dados que não apresentam padrão de comportamento. Imagina-se por exemplo o passeio de uma grão de polén na natureza. Seu comportamento ao percorrer o mundo real não tem qualquer padrão, visto que é influencidao pelas pelo movimento molecular de um gota d'agua que é aleatório. (ver mais em Curso Intensivo de Python)

```python
# Classe RandomWalk()

from random import choice

class RandomWalk():
    def __init__(self, num_points=5000):
        self.num_points = num_points
        
        # Todos os passeio devem começar em 0,0
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        while len(self.x_values) < self.num_points:
            
            # decide a direção a ser seguida e distância a ser percorrida nessa direção
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # rejeita movimentos que não vão a lugar nenhum
            if x_step == 0 and y_step == 0:
                continue
            
            #Calcula os próximos valores de x e de y
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)
```

```python
# Plotando os passeio

import matplotlib.pyplot as plt

#Cria um passeio aleatório
rw = RandomWalk()
rw.fill_walk()
```

```python
plt.scatter(rw.x_values, rw.y_values, s=15)
plt.show()
```

```python
## Criando vários passeios

while True:
    rw = RandomWalk()
    rw.fill_walk()
    plt.scatter(rw.x_values, rw.y_values, s=15)
    plt.show()
    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```
