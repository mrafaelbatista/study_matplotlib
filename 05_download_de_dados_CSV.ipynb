## PLOTANDO UMA SEGUNDA SÉRIE DE DADOS

```python
# plotando um gráfico de temperatura PERÍODO DE TEMPO MAIOR

import csv
from datetime import datetime
from matplotlib import pyplot as plt

filename = 'chapter_16/sitka_weather_2014.csv' 

# Obtendo as DATAS e a temperaturas máximas do arquivo
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[1])
        highs.append(high)
        
        low = int(row[3])
        lows.append(low)
        
# plotagem dos dados
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red')
plt.plot(dates, lows, c='blue')

# formatando o gráfico
plt.title("Dialy high and lows temperatures - 2014", fontsize=18)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```

## Sombreando uma área do gráfico

```python
# plotando um gráfico de temperatura PERÍODO DE TEMPO MAIOR

import csv
from datetime import datetime
from matplotlib import pyplot as plt

filename = 'chapter_16/sitka_weather_2014.csv' 

# Obtendo as DATAS e a temperaturas máximas do arquivo
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[0], "%Y-%m-%d")
        dates.append(current_date)
        
        high = int(row[1])
        highs.append(high)
        
        low = int(row[3])
        lows.append(low)
        
# plotagem dos dados
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red', alpha=0.5)
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# formatando o gráfico
plt.title("Dialy high and lows temperatures - 2014", fontsize=18)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```

## Verificação de erros

```python
# plotando um gráfico de temperatura PERÍODO DE TEMPO MAIOR

import csv
from datetime import datetime
from matplotlib import pyplot as plt

filename = 'chapter_16/death_valley_2014.csv' 

# Obtendo as DATAS e a temperaturas máximas do arquivo
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    dates, highs, lows = [], [], []
    for row in reader:
        try:
            current_date = datetime.strptime(row[0], "%Y-%m-%d")
            high = int(row[1])
            low = int(row[3])        
        except ValueError:
            print(current_date, 'missing data')
        else:
            dates.append(current_date)
            highs.append(high)
            lows.append(low)
        
# plotagem dos dados
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs, c='red', alpha=0.5)
plt.plot(dates, lows, c='blue', alpha=0.5)
plt.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# formatando o gráfico
plt.title("Dialy high and lows temperatures - 2014 \nDeath Valley, CA", fontsize=18)
plt.xlabel('', fontsize=12)
fig.autofmt_xdate()
plt.ylabel("Temperature (F)", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=12)

plt.show()
```
